```python
#create coord file for cp2k

def create_coord(coord):
    count = 0
    coord_vir = []
    for i in range(len(coord)):
        s = coord[i].split()
        if len(s)>=1 and s[0]=='O':
            coord_O = [float(s[1]),float(s[2]),float(s[3])]
            coord_H_1 = [float(coord[i+1].split()[1]),float(coord[i+1].split()[2]),float(coord[i+1].split()[3])]
            coord_H_2 = [float(coord[i+2].split()[1]),float(coord[i+2].split()[2]),float(coord[i+2].split()[3])]
            vs = add_vir(coord_O,coord_H_1,coord_H_2)
            temp_O = "O" + ' '+str(coord_O[0])+' '+str(coord_O[1])+' '+str(coord_O[2])+' '+'H2O'+'\n'
            temp_H_1 = "H" + ' '+str(coord_H_1[0])+' '+str(coord_H_1[1])+' '+str(coord_H_1[2])+' '+'H2O'+'\n'
            temp_H_2 = "H" + ' '+str(coord_H_2[0])+' '+str(coord_H_2[1])+' '+str(coord_H_2[2])+' '+'H2O'+'\n'
            temp_M = "M" + ' '+str(vs[0])+' '+str(vs[1])+' '+str(vs[2])+' '+'H2O'+'\n'
            coord_vir.append(temp_O)
            coord_vir.append(temp_H_1)
            coord_vir.append(temp_H_2)
            coord_vir.append(temp_M)
            count = count+4
        elif s[0]=='Na' or s[0]=='Cl' or s[0]=='La':
            count = count+1
            coord_vir.append(coord[i])
            continue
    coord_vir_wrap = wrap_coord(coord_vir)
    return coord_vir,coord_vir_wrap,count
```

```python
##add virtual particle for DC water

def add_vir(coord_O,coord_H_1,coord_H_2):
    dist = 0.215              #distance between O and virtual site
    x_O = coord_O[0]          #extract coordinates
    y_O = coord_O[1]
    z_O = coord_O[2]
    x_H_1 = coord_H_1[0]
    y_H_1 = coord_H_1[1]
    z_H_1 = coord_H_1[2]
    x_H_2 = coord_H_2[0]
    y_H_2 = coord_H_2[1]
    z_H_2 = coord_H_2[2]
    
    v1 = np.array([x_H_1-x_O,y_H_1-y_O,z_H_1-z_O])
    v2 = np.array([x_H_2-x_O,y_H_2-y_O,z_H_2-z_O])
    v1 = v1/(np.dot(v1,v1))**0.5
    v2 = v2/(np.dot(v2,v2))**0.5
    v_bis = v1+v2
    bas_1 = v_bis/(np.dot(v_bis,v_bis))**0.5
    pos_vec = dist * bas_1          #position of virtual site
    vir_site = pos_vec + coord_O
    return vir_site
```

```python
##wrap coordinates

def wrap_coord(coord_vir):
    O = []
    H = []
    box_len = 14.2784                #length of simulation box
    h_box_len = box_len/2
    coord_vir_wrap = []

    for i in range(len(coord_vir)):
        s = coord_vir[i].split()
        x = float(coord_vir[i].split()[1])
        y = float(coord_vir[i].split()[2])
        z = float(coord_vir[i].split()[3])
        while x>h_box_len:
            x = x - box_len
        while x<-h_box_len:
            x = x + box_len
        while y>h_box_len:
            y = y - box_len
        while y<-h_box_len:
            y = y + box_len
        while z>h_box_len:
            z = z - box_len
        while z<-h_box_len:
            z = z + box_len
        temp = [s[0],str(x),str(y),str(z)]
        if s[0]=='O' or s[0]=='H' or s[0]=='M':
            temp = '  '.join(temp)+' '+'H2O'+'\n'
            coord_vir_wrap.append(temp)
        else:
            temp = '  '.join(temp)+'\n'
            coord_vir_wrap.append(temp)  
    return coord_vir_wrap
```

```python
#calculate permanent dipole moment of water molecule

def get_permanent_cmd(coord_vir,dp_cmd):
    dp_total_cmd = []
    k = 0
    for i in range(len(coord_vir)):
        s = coord_vir[i].split()
        if s[0]=='O':
            H1_x = float(coord_vir[i+1].split()[1])
            H1_y = float(coord_vir[i+1].split()[2])
            H1_z = float(coord_vir[i+1].split()[3])
            H2_x = float(coord_vir[i+2].split()[1])
            H2_y = float(coord_vir[i+2].split()[2])
            H2_z = float(coord_vir[i+2].split()[3])
            M_x = float(coord_vir[i+3].split()[1])
            M_y = float(coord_vir[i+3].split()[2])
            M_z = float(coord_vir[i+3].split()[3])
            temp_dp_x = (-1.0380*(M_x))+(0.5190*(H1_x+H2_x))
            temp_dp_y = (-1.0380*(M_y))+(0.5190*(H1_y+H2_y))
            temp_dp_z = (-1.0380*(M_z))+(0.5190*(H1_z+H2_z))
            dp_total_cmd.append(((dp_cmd[k][0]+temp_dp_x),(dp_cmd[k][1]+temp_dp_y),\
                                 (dp_cmd[k][2]+temp_dp_z)))
            k = k+1
        elif s[0]=='Cl' or s[0]=='Na' or s[0]=='La':
            dp_total_cmd.append((dp_cmd[k][0],dp_cmd[k][1],dp_cmd[k][2]))
            k = k+1
    return dp_total_cmd
```

```python
def get_dipole_aimd(coord_uw):
    count = 0
    count_O = 0
    dp_aimd = []
    dp_x_ion_tot = 0
    dp_y_ion_tot = 0
    dp_z_ion_tot = 0
    for i in range(len(coord_uw)):
        s = coord_uw[i].split()
        if s[0]=='X':
            count = count + 1
            if count ==1:
                start = i
    for i in range(len(coord_uw)):
        s = coord_uw[i].split()
        if s[0]=='O':
            #count_O = count_O + 1
            #print('Found',count_O)
            s1 = coord_uw[i+1].split()
            s2 = coord_uw[i+2].split()
            s3 = coord_uw[start].split()
            s4 = coord_uw[start+1].split()
            s5 = coord_uw[start+2].split()
            s6 = coord_uw[start+3].split()
            
            dp_x_ion = 6*float(s[1]) + float(s1[1]) + float(s2[1])
            dp_y_ion = 6*float(s[2]) + float(s1[2]) + float(s2[2])
            dp_z_ion = 6*float(s[3]) + float(s1[3]) + float(s2[3])
        
            #print(coord[start_wan])
            dp_x_wan = -2*(float(s3[1]) + float(s4[1]) + float(s5[1]) + float(s6[1]))
            dp_y_wan = -2*(float(s3[2]) + float(s4[2]) + float(s5[2]) + float(s6[2]))
            dp_z_wan = -2*(float(s3[3]) + float(s4[3]) + float(s5[3]) + float(s6[3]))
            
            dp_x_tot =  dp_x_ion + dp_x_wan
            dp_y_tot =  dp_y_ion + dp_y_wan
            dp_z_tot = dp_z_ion + dp_z_wan
            dp_aimd.append((dp_x_tot,dp_y_tot,dp_z_tot))
            start = start + 4
        elif s[0]=='H':
            continue
        elif s[0]=='X':
            continue
        elif s[0]=='La':
            #print('Found',s[0])
            s3 = coord_uw[start].split()
            s4 = coord_uw[start+1].split()
            s5 = coord_uw[start+2].split()
            s6 = coord_uw[start+3].split()
            
            dp_x_ion = float(s[1])
            dp_y_ion = float(s[2])
            dp_z_ion = float(s[3])
        
            #calculate dipole moment from Wannier centers
            dp_x_wan = 8*(dp_x_ion) - 2*(float(s3[1])+float(s4[1])+float(s5[1])+float(s6[1]))
            dp_y_wan = 8*(dp_y_ion) - 2*(float(s3[2])+float(s4[2])+float(s5[2])+float(s6[2]))
            dp_z_wan = 8*(dp_z_ion) - 2*(float(s3[3])+float(s4[3])+float(s5[3])+float(s6[3]))
            dp_aimd.append((dp_x_wan,dp_y_wan,dp_z_wan))
            start = start + 4
        elif s[0]=='Cl':
            #print('Found',s[0])
            s3 = coord_uw[start].split()
            s4 = coord_uw[start+1].split()
            s5 = coord_uw[start+2].split()
            s6 = coord_uw[start+3].split()
            
            dp_x_ion = float(s[1])
            dp_y_ion = float(s[2])
            dp_z_ion = float(s[3])
        
            dp_x_wan = -2*((float(s3[1])-dp_x_ion) + (float(s4[1])-dp_x_ion) + (float(s5[1])-dp_x_ion) + (float(s6[1])-dp_x_ion))
            dp_y_wan = -2*((float(s3[2])-dp_y_ion) + (float(s4[2])-dp_y_ion) + (float(s5[2])-dp_y_ion) + (float(s6[2])-dp_y_ion))
            dp_z_wan = -2*((float(s3[3])-dp_z_ion) + (float(s4[3])-dp_z_ion) + (float(s5[3])-dp_z_ion) + (float(s6[3])-dp_z_ion))
            
            dp_aimd.append((dp_x_wan,dp_y_wan,dp_z_wan))
            start = start + 4  
    return dp_aimd
```

```python
def cost_func(params):
    error = 0
    conv = 0.529177
    count_fr = 0
    option = 1
    nframe = 110     #number of frames
    count_int = 253  #number of atoms for which dipole is calculated in each frame
    
    filename_aimd = '/home/rajorshi/Desktop/Pol_MD/Cl_water_dp_5kbar/converged_data_big/Cl_water_wannier.xyz' # coord file with wrapped coordinates
    with open(filename_aimd) as f1:
        lines_aimd = f1.readlines()
        
    filename_aimd_uw = '/home/rajorshi/Desktop/Pol_MD/Cl_water_dp_5kbar/converged_data_big/Cl_water_wannier_uw.xyz' # coord file with unwrapped coordinates for dipole moment calc
    with open(filename_aimd_uw) as f1_uw:
        lines_aimd_uw = f1_uw.readlines()
    for i in range(len(lines_aimd_uw)):
        s = lines_aimd_uw[i].split()
        if len(s)==1:
            count_fr = count_fr+1
            print('Frame',count_fr)
            n_atoms = int(s[0])
            spe = []
            coord = []
            dp_cmd = []
            tot_x = 0
            tot_y = 0
            tot_z = 0
            flag = 0
            j = i+2
            while flag!= 1:
                if lines_aimd_uw[j].split()[0] != 'X':
                    spe.append(lines_aimd_uw[j].split()[0])
                    coord.append(lines_aimd_uw[j])
                    j = j+1
                else:
                    flag = 1
            
            coord_uw = lines_aimd_uw[i+2:i+2+n_atoms]
            coord_vir,coord_vir_wrap,count = create_coord(coord)
        
            if option == 1:
                create_inp_1(params,coord_vir,count)
            else:
                create_inp_2(params,coord_vir,count)
            
            subprocess.run(["/home/rajorshi/Desktop/Pol_MD/Cl_water_dp_5kbar/myscript.sh"]) #run_cp2k
            filename = '/home/rajorshi/Python/BM_potential/multipoles'                      #classical dipoles
            
            with open(filename) as f:
                lines_cmd = f.readlines()
            k = 0
            for j in range(len(spe)):
                if spe[j] == 'O':
                    temp_x = float(lines_cmd[k].split()[0])+float(lines_cmd[k+1].split()[0])+float(lines_cmd[k+2].split()[0])+float(lines_cmd[k+3].split()[0])
                    temp_y = float(lines_cmd[k].split()[1])+float(lines_cmd[k+1].split()[1])+float(lines_cmd[k+2].split()[1])+float(lines_cmd[k+3].split()[1])
                    temp_z = float(lines_cmd[k].split()[2])+float(lines_cmd[k+1].split()[2])+float(lines_cmd[k+2].split()[2])+float(lines_cmd[k+3].split()[2])
                    dp_cmd.append((temp_x*conv,temp_y*conv,temp_z*conv))   #induced dipole from cmd
                    k = k+4
                elif spe[j] == 'H':
                    continue
                
                else:
                    temp_x = float(lines_cmd[k].split()[0])
                    temp_y = float(lines_cmd[k].split()[1])
                    temp_z = float(lines_cmd[k].split()[2])
                    tot_x = tot_x + temp_x
                    tot_y = tot_y + temp_y
                    tot_z = tot_z + temp_z
                    dp_cmd.append((temp_x*conv,temp_y*conv,temp_z*conv))
                    k = k + 1
            dp_aimd = get_dipole_aimd(coord_uw)
            dp_total_cmd = get_permanent_cmd(coord_vir,dp_cmd)  #induced+permanent dipole moment

            k = 0
            for j in range(len(spe)):
                if spe[j] == 'O':
                    temp = (dp_aimd[k][0] - dp_total_cmd[k][0])**2 + \
                       (dp_aimd[k][1] - dp_total_cmd[k][1])**2 +  \
                       (dp_aimd[k][2] - dp_total_cmd[k][2])**2
                    temp = temp/((dp_aimd[k][0]**2)+(dp_aimd[k][1]**2)+(dp_aimd[k][2]**2))
                    error = (error + temp)                      #activate if water dp is needed for La-O,Cl-O
                    k = k + 1
                elif spe[j] == 'H':
                    continue
                else:
                    temp = (dp_aimd[k][0] - dp_total_cmd[k][0])**2 + \
                       (dp_aimd[k][1] - dp_total_cmd[k][1])**2 +  \
                       (dp_aimd[k][2] - dp_total_cmd[k][2])**2
                    temp = temp/((dp_aimd[k][0]**2)+(dp_aimd[k][1]**2)+(dp_aimd[k][2]**2))
                    error = (error + temp)
                    k = k + 1
    return (error/(nframe*count_int)) 
```
